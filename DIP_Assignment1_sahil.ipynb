### Importing modules

```python
import PIL #pillow
from PIL import Image , ImageOps ,ImageColor, ImageFilter

import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

## <font color = 'red'>Take your passport size photograph and apply following transformations on it:</font>

```python
Photo = Image.open("Images/Photo.jpg")

fig = plt.figure(figsize=(5,5))
plt.title('Photo')
plt.axis("off")
implot = plt.imshow(Photo )
```

## a. Convert Image in gray scale

```python
gray_img = Photo.convert('L')

plt.figure(figsize=(7,7))
plt.subplot(1, 2, 1) 
plt.title('Original image')
plt.axis("off")
implot = plt.imshow(Photo)
    
plt.subplot(1, 2, 2) 
plt.title('Gray image')
plt.axis("off")
implot = plt.imshow(gray_img , cmap='gray', vmin=0, vmax=255)
```

## b. Convert image in black and white (binary format) 

```python
thresh = 128
binary_img = gray_img.point(lambda p: p > thresh and 255)

plt.figure(figsize=(7,7))
plt.subplot(1, 2, 1)
plt.title('Gray image')
plt.axis("off")
implot = plt.imshow(gray_img , cmap='gray', vmin=0, vmax=255)
    
plt.subplot(1, 2, 2) 
plt.title('Black and white image')
plt.axis("off")
implot = plt.imshow(binary_img , cmap='gray', vmin=0, vmax=255)
```

## c. Average filter on gray scale

```python
##Two smoothing (averaging) filter masks
smooth_kernel1 = np.array(( 
            (1,1,1),
            (1,1,1),
            (1,1,1),
            )) /9

smooth_kernel2 = np.array((
            (1,2,1),
            (2,4,2),
            (1,2,1),
            )) /16


# We are taking Kernel Size as 3x3
avg_filter1 = gray_img.filter(ImageFilter.Kernel((3, 3), smooth_kernel1.flatten(), 1, 0))

avg_filter2 = gray_img.filter(ImageFilter.Kernel((3, 3), smooth_kernel2.flatten(), 1, 0))

plt.figure(figsize=(10,10))
plt.subplot(1, 3, 1) 
plt.title('Gray image')
plt.axis("off")
implot = plt.imshow(gray_img , cmap='gray', vmin=0, vmax=255)
    
plt.subplot(1, 3, 2) 
plt.title('Averaging with kernel 1')
plt.axis("off")
implot = plt.imshow(avg_filter1 , cmap='gray', vmin=0, vmax=255)

plt.subplot(1, 3, 3) 
plt.title('Averaging with kernel 2')
plt.axis("off")
implot = plt.imshow(avg_filter2 , cmap='gray', vmin=0, vmax=255)
```

## h.	Logarithm transformation on gray scale

```python
def logTransformImg(img , c = None):
    '''function returns log transformed image'''
    if(c == None):
        c = 255/(np.log(1 + np.max(img)))#defualt value of C

    log_transformed = np.round(c * np.log(1 + np.array(img))) #log_transformed is only array 
    return Image.fromarray(np.array(log_transformed, int)) # converting to int and image
 


log_transformed_img1 = logTransformImg(gray_img)
log_transformed_img2 = logTransformImg(gray_img , 25)

plt.figure(figsize=(10,10))
plt.subplot(1, 3, 1) 
plt.title('Gray image')
plt.axis("off")
implot = plt.imshow(gray_img , cmap='gray', vmin=0, vmax=255)
    
plt.subplot(1, 3, 2) 
plt.title('Log Transformed C = defualt ')
plt.axis("off")
implot = plt.imshow(log_transformed_img1 , cmap='gray', vmin=0, vmax=255)

plt.subplot(1, 3, 3)
plt.title('Log Transformed C = 25')
plt.axis("off")
implot = plt.imshow(log_transformed_img2 , cmap='gray', vmin=0, vmax=255)
```

## i.	Power law transformation on gray scale

```python
def gamma_corrected(img ,gamma ,c = 1):
    '''function returns gamma corrected image'''
    gamma_corrected_img =  c*(255*(np.array(img) / 255) ** gamma) # adjustment for intensities
    return Image.fromarray(np.array(gamma_corrected_img, int))

gamma_corrected_img0 = gamma_corrected(gray_img , 0.2)
gamma_corrected_img1 = gamma_corrected(gray_img , 2.5)
gamma_corrected_img2 = gamma_corrected(gray_img , 2.5 , 3)

plt.figure(figsize=(13,13))
plt.subplot(1, 4, 1) 
plt.title('Gray image')
plt.axis("off")
implot = plt.imshow(gray_img , cmap='gray', vmin=0, vmax=255)

plt.subplot(1, 4, 2) 
plt.title('gamma corrected \ngamma = 0.2, c = 1 ')
plt.axis("off")
implot = plt.imshow(gamma_corrected_img0 , cmap='gray', vmin=0, vmax=255)
    
plt.subplot(1, 4, 3) 
plt.title('gamma corrected \ngamma = 2.5, c = 1 ')
plt.axis("off")
implot = plt.imshow(gamma_corrected_img1 , cmap='gray', vmin=0, vmax=255)

plt.subplot(1, 4, 4)
plt.title('gamma corrected \ngamma = 2.5, c = 3')
plt.axis("off")
implot = plt.imshow(gamma_corrected_img2 , cmap='gray', vmin=0, vmax=255)
```
