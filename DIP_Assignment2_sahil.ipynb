### Importing modules

```python
import PIL #pillow
from PIL import Image , ImageOps ,ImageColor, ImageFilter

import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

## <font color = 'red'>For the following image apply the following steps:</font>

```python
image = Image.open("Images/Colorcircle.jpg")

plt.figure(figsize=(5,5))
plt.title('Photo')
implot = plt.imshow(image)
```

## a.	Obtain histogram of the image

```python

r, g, b = image.split() 

plt.figure(figsize = (17,4))

plt.subplot(1,3,1)
plt.title('Red')
plt.hist(np.array(r).flatten() , color = "red",bins = range(256) )

plt.subplot(1,3,2)
plt.title('Green')
plt.hist(np.array(g).flatten() , color = "green",bins = range(256) )

plt.subplot(1,3,3)
plt.title('Blue')
plt.hist(np.array(b).flatten() , color = "blue",bins = range(256) )

plt.show()
```

#### <font color = 'green' >In all the above histograms no of pixels having intensity 255 is highest because background of circular image is white </font>

```python
r, g, b = image.split() 

plt.figure(figsize = (17,4))

plt.subplot(1,3,1)
plt.title('Red')
plt.hist(np.array(r).flatten() , color = "red" ,bins = range(255) )

plt.subplot(1,3,2)
plt.title('Green')
plt.hist(np.array(g).flatten() , color = "green",bins = range(255) )

plt.subplot(1,3,3)
plt.title('Blue')
plt.hist(np.array(b).flatten() , color = "blue",bins = range(255) )

plt.show()
```

#### <font color = 'green'> In these plots to overcome above problem We have excluded intensity value 255 </font> 

## b.	Convert image into gray scale and obtain histogram of the image

```python
gray_image = image.convert('L')

plt.figure(figsize=(10,10))
plt.subplot(1, 2, 1) 
plt.title('Original image')
implot = plt.imshow(image)
    
plt.subplot(1, 2, 2) 
plt.title('Gray image')
implot = plt.imshow(gray_image , cmap='gray', vmin=0, vmax=255)

plt.figure(figsize =(10, 7)) 
plt.hist(np.array(gray_image).flatten(),bins = range(256) , color = "black")   
plt.title("Histogram of gray scaled image")    
plt.show()

plt.figure(figsize =(10, 7)) 
plt.hist(np.array(gray_image).flatten(),bins = range(255) , color = "black")   
plt.title("Histogram of gray scaled image Excluding intensity 255 because background is white")    
plt.show()

```

## c.	Convert image into binary and obtain histogram of the image

```python
thresh = 120
binary_image = gray_image.point(lambda p: p > thresh and 255)

plt.figure(figsize=(10,10))

plt.subplot(1, 2, 1) 
plt.title('Original image')
implot = plt.imshow(image)
    
plt.subplot(1, 2, 2) 
plt.title('Binary image')
implot = plt.imshow(binary_image , cmap='gray', vmin=0, vmax=255)

plt.figure(figsize =(10, 7)) 
plt.hist(np.array(binary_image).flatten(),bins = range(256) , color = "black")   
plt.title("Histogram of binary image")    
plt.show() 
```
